```python
!pip install easyocr
import easyocr
import gdown

# Создаем необходимые директории
os.makedirs('/content/EasyOCR/user_network', exist_ok=True)
os.makedirs('/content/EasyOCR/model/', exist_ok=True)

# Загрузка yaml-файла для обученной модели
custom_yaml_link = 'https://drive.google.com/file/d/14DLULWAkr4qXX1YnsoN5vK-nNk9Cwbz5/view?usp=sharing'
custom_yaml_file = '/content/EasyOCR/user_network/custom_ocr.yaml'
gdown.download(custom_yaml_link, custom_yaml_file, fuzzy=True, quiet=True)

# Загрузка файла custom_ocr.py для обученной модели
custom_py_link = 'https://drive.google.com/file/d/1dP1OepFEaLEI2YJkz0HQZzijIkhNhqVX/view?usp=sharing'
custom_py_file = '/content/EasyOCR/user_network/custom_ocr.py'
gdown.download(custom_py_link, custom_py_file, fuzzy=True, quiet=True)

# Загрузка обученной модели (ссылка будет обновляться в связи с дообучением модели!)
custom_model_link = 'https://drive.google.com/file/d/1ah6UJnx6WSpv5l9trVydHl7l-qa_jhio/view?usp=sharing' # best_accuracy4
custom_model_file = '/content/EasyOCR/model/custom_ocr.pth'
gdown.download(custom_model_link, custom_model_file, fuzzy=True, quiet=True)

# Список символов, которые будут выводиться
allow_list = "0123456789!\"#$%&()*+,-./:;<=>?@[\\]№_{}~ €₽«»°ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyzАБВГДЕЁЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯабвгдеёжзийклмнопрстуфхцчшщъыьэюя"

def ocr_process(image):
    ''' Функция распознавания текста на ч\б изображении
    выход: все распознанные слова через пробел
    '''

    reader = easyocr.Reader(['en'],
                            model_storage_directory='EasyOCR/model',
                            user_network_directory='EasyOCR/user_network',
                            recog_network='custom_ocr')
    result = reader.readtext(image,
                             detail = 0,                 # не сохраняем координаты при распознавании
                             allowlist=allow_list,       # список выводимых символов
                             )

    # Формирование финального текста
    words = ' '.join(result)

    return words
```
